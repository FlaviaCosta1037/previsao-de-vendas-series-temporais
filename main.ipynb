```python
import pandas as pd
from src.data_loader import carregar_dados
from src.data_clean import tratar_dados
from src.grouped import agrupar_serie
from src.eda import plot_evolucao_vendas
from src.outliers import detectar_outliers, plot_outliers
from src.decomposition import decompor_serie
from src.correlation import heatmap_correlacao
from src.autocorrelacao import autocorrelacionar
from src.sarimax_model import modelar_sarimax
from src.svr_model import modelar_svr
from src.teste_adf_pacf import testar_adf_pacf
from src.divisao_treino_teste_validacao import dividir_modelagem
from src.frases import separador, aplicar_frase
from src.arima_model import modelar_arima_automatico, modelar_arima_manual

CAMINHO = 'data/0+12 2025 - LatAm Business Analytics_No links.xlsx'

df = carregar_dados(CAMINHO)
df = tratar_dados(df)
df_agrupado = agrupar_serie(df)
plot_evolucao_vendas(df_agrupado)

outliers, lim_inf, lim_sup = detectar_outliers(df_agrupado, 'Units')
plot_outliers(df_agrupado, 'Units', lim_inf, lim_sup)
decompor_serie(df_agrupado['Units'])
heatmap_correlacao(df)
autocorrelacionar(df_agrupado)
testar_adf_pacf(df_agrupado)

separador()
aplicar_frase('Informação do tamanho do treino, teste e validação')
dividir_modelagem(df_agrupado)

separador()
aplicar_frase("Resultado ARIMA Automático")

resultados_auto = modelar_arima_automatico(df_agrupado)

print(f"\nOrdem final encontrada: {resultados_auto['melhor_ordem']}")
print(f"MAPE Validação: {resultados_auto['mape_val']:.2f}%")
print(f"MAPE Teste: {resultados_auto['mape_test']:.2f}%")

separador()
aplicar_frase('Resultado ARIMA manual')
entrada = input("Digite a ordem (p, d, q) separada por vírgulas: ")
# Exemplo de entrada: 5,1,2
p, d, q = map(int, entrada.split(','))
ordem_escolhida = (p, d, q)
resultados = modelar_arima_manual(df_agrupado, ordem_escolhida)

separador()
aplicar_frase("Resultado SARIMAX com normalização Scaled")

resultados_sarimax = modelar_sarimax(df_agrupado, auto=True, exog_type='scale')

print(f"Ordem final: {resultados_sarimax['order']}")
print(f"Sazonalidade: {resultados_sarimax['seasonal_order']}")
print(f"MAPE Validação: {resultados_sarimax['mape_val']:.2f}%")
print(f"MAPE Teste: {resultados_sarimax['mape_test']:.2f}%")

separador()
aplicar_frase("Resultado SARIMAX com normalização LOG")

resultados_sarimax = modelar_sarimax(df_agrupado, auto=True, exog_type='log')

print(f"Ordem final: {resultados_sarimax['order']}")
print(f"Sazonalidade: {resultados_sarimax['seasonal_order']}")
print(f"MAPE Validação: {resultados_sarimax['mape_val']:.2f}%")
print(f"MAPE Teste: {resultados_sarimax['mape_test']:.2f}%")

separador()
aplicar_frase("Resultado SVR sem exógena")
modelar_svr(df_agrupado, exog_col=None, stepahead=3, dimension=12, C=100, epsilon=0.01, gamma=0.001)

separador()
aplicar_frase("Resultado SVR com exógena")
modelar_svr(df_agrupado, exog_col='Commission USD', stepahead=3, dimension=12, C=100, epsilon=0.01, gamma=0.001)


```
